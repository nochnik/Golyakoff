```python
def product_parser(item):    
    page_number = 1
    category_name = list(item.keys())[0]
    base_url = item[category_name]
    # loop through all pages until there are no more pages left
    while True:
        url = base_url + '?PAGEN_1=' + str(page_number)
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        products = soup.find_all('a', href=lambda href: href and href.startswith('/catalog/product/'))

        for product in products:
            if(product.span != None):
                product_name = product.span.text.strip()
                if product_name and not product_name.startswith('\n'):
                    list_of_bobs.append({category_name : product_name})

        next_page_link = soup.find('div', {'class' :"ajax_load_btn rounded3 colored_theme_hover_bg"})
        if not next_page_link:
            break
        
        page_number += 1
```

```python
for i in list_of_urls:
    product_parser(i)
```
